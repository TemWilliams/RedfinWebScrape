```python
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
import pandas

r=requests.get("https://www.redfin.com/city/1826/MA/Boston", headers={'User-agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:61.0) Gecko/20100101 Firefox/61.0'})
c=r.content

soup=BeautifulSoup(c, "html.parser")
```

```python
all=soup.find_all("div",{"class":"bottomV2"})
```

```python
l=[]
for item in all:
    d={}
    d["Price"]=item.find("span",{"class":"homecardV2Price"}).text
    d["Address"]=item.find("div", {"class":"homeAddressV2"}).text
    d["Home Info"]=item.find("div", {"class":"HomeStatsV2 font-size-small"}).text
    print(" ")
    l.append(d)
```

```python
l
```

```python
df=pandas.DataFrame(l)
```
